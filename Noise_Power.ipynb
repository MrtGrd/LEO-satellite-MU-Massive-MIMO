```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as mpl
# Set font size for the plot
mpl.rcParams['font.family'] = 'DejaVu Serif'
mpl.rcParams['font.size'] = 10  # Base font size
mpl.rcParams['axes.titlesize'] = 10  # Title font size
mpl.rcParams['axes.labelsize'] = 10  # Axis label font size
mpl.rcParams['xtick.labelsize'] = 10  # X-axis tick label size
mpl.rcParams['ytick.labelsize'] = 10  # Y-axis tick label size
mpl.rcParams['legend.fontsize'] = 10  # Legend font size

# Constants
k = 1.38e-23  # Boltzmann constant in J/K
T = 290  # Temperature in Kelvin
K = 10 * np.log10(k * T) + 30  # Thermal noise power density constant in dBm/Hz (includes factor to convert from Watts to dBm)

# Define bandwidth range (from 0.1 MHz to 100 MHz)
bw_range = np.logspace(np.log10(0.1e6), np.log10(100e6), 100)  # From 0.1 MHz to 100 MHz (logarithmic scale)

# Calculate noise power (Pnoise) for each bandwidth in dBm
P_noise = K + 10 * np.log10(bw_range)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(bw_range / 1e6, P_noise, label="Noise Power (Pnoise)", color='b')  # Convert bandwidth to MHz
plt.xscale('log')  # Logarithmic scale for x-axis (bandwidth)
plt.yscale('linear')  # Linear scale for y-axis (dBm)
plt.title("Noise Power vs Bandwidth in MHz")
plt.xlabel("Bandwidth (MHz)")  # Label x-axis in MHz
plt.ylabel("Noise Power [dBm]")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.show()
```
