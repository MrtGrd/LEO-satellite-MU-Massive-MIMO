```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.ticker as ticker
import matplotlib as mpl

# Customize matplotlib settings
mpl.rcParams['font.family'] = 'DejaVu Serif'
mpl.rcParams['font.size'] = 10  # Base font size
mpl.rcParams['axes.titlesize'] = 12  # Title font size
mpl.rcParams['axes.labelsize'] = 10  # Axis label font size
mpl.rcParams['xtick.labelsize'] = 10  # X-axis tick label size
mpl.rcParams['ytick.labelsize'] = 10  # Y-axis tick label size
mpl.rcParams['legend.fontsize'] = 10  # Legend font size

# Define filter parameters
n = 8  # Filter order
f_center = 100e6  # Center frequency in Hz (150 MHz)
BW = 50e6  # Bandwidth in Hz (50 MHz)
ripple_dB = 1  # Passband ripple in dB
attenuation_dB = 96  # Stopband attenuation in dB
f_s = 400e6  # 400 MHz sampling frequency

# Design the elliptic bandpass filter
f_low = f_center - BW / 2  # Lower cutoff frequency
f_high = f_center + BW / 2  # Upper cutoff frequency

# Normalize frequencies by the Nyquist frequency (f_s / 2)
low = f_low / (f_s / 2)
high = f_high / (f_s / 2)

# Ensure normalized frequencies are valid
if not (0 < low < 1) or not (0 < high < 1):
    raise ValueError("Cutoff frequencies must satisfy 0 < f_low, f_high < f_s / 2.")

# Design the elliptic filter using scipy.signal
b, a = signal.ellip(n, ripple_dB, attenuation_dB, [low, high], btype='bandpass', analog=False)

# Print the filter coefficients with 3 decimal points
print("Filter coefficients (b):")
for coefficient in b:
    print(f"{coefficient:.3f}")

print("\nFilter coefficients (a):")
for coefficient in a:
    print(f"{coefficient:.3f}")

# Frequency response calculation
w, h = signal.freqz(b, a, worN=8000)
frequencies = w / np.pi * (f_s / 2)  # Convert to Hz
frequencies_MHz = frequencies / 1e6  # Convert to MHz for display

# Frequency response calculation
w, h = signal.freqz(b, a, worN=8000)
frequencies = w / np.pi * (f_s / 2)  # Convert to Hz
frequencies_MHz = frequencies / 1e6  # Convert to MHz for display

# Plot frequency response
plt.figure(figsize=(10, 6))
plt.semilogx(frequencies_MHz, 20 * np.log10(abs(h)), label=f'{n}-th degree Elliptic Filter', linewidth=2)

# Mark the lower and upper cutoff frequencies in MHz
plt.axvline(f_low / 1e6, color='r', linestyle='--', label=f'Lower cutoff: {f_low / 1e6} MHz')
plt.axvline(f_high / 1e6, color='r', linestyle='--', label=f'Upper cutoff: {f_high / 1e6} MHz')

# Limit the frequency span on the plot
plt.xlim(50, 250)  # Frequency axis in MHz

# Labels and title
plt.title(f'Frequency Response of {n}-th Degree Elliptic Bandpass Filter')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True, which="both", ls="--")
plt.legend(loc='best')

# Use ScalarFormatter to avoid scientific notation on the x-axis
formatter = ticker.ScalarFormatter()
formatter.set_scientific(False)
formatter.set_useOffset(False)

# Apply the formatter to the x-axis
plt.gca().xaxis.set_major_formatter(formatter)

# Customize the x-axis ticks to show in MHz without scientific notation
plt.xticks(np.arange(20, 251, 25))  # Custom ticks in MHz, from 50 MHz to 250 MHz

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.ticker as ticker

# Define the filter parameters
n = 6  # Filter order
f_center = 100e6  # Center frequency in Hz
BW = 100e6  # Bandwidth in Hz
ripple_dB = 1  # Passband ripple in dB
attenuation_dB = 96  # Stopband attenuation in dB
f_s = 400e6  # 400 MHz sampling frequency

# Design the elliptic bandpass filter
f_low = f_center - BW / 2  # Lower cutoff frequency
f_high = f_center + BW / 2  # Upper cutoff frequency

# Normalize frequencies by the Nyquist frequency (f_s / 2)
low = f_low / (f_s / 2)
high = f_high / (f_s / 2)

# Design the elliptic filter using scipy.signal
b, a = signal.ellip(n, ripple_dB, attenuation_dB, [low, high], btype='bandpass', analog=False)

# Function to apply perturbations
def apply_perturbation(coefficients, tolerance=0.01):
    """Apply a random perturbation to the filter coefficients."""
    return coefficients * (1 + np.random.uniform(-tolerance, tolerance, size=coefficients.shape))

# Create perturbations and calculate new responses
b_perturbed = apply_perturbation(b)
a_perturbed = apply_perturbation(a)

# Frequency response calculation
w, h_original = signal.freqz(b, a, worN=8000)
_, h_perturbed = signal.freqz(b_perturbed, a_perturbed, worN=8000)

frequencies = w / np.pi * (f_s / 2)  # Convert to Hz
frequencies_MHz = frequencies / 1e6  # Convert to MHz for display

# Plot frequency response
plt.figure(figsize=(10, 6))

# Plot original and perturbed frequency response
plt.semilogx(frequencies_MHz, 20 * np.log10(abs(h_original)), label='Original Filter', linewidth=2)
plt.semilogx(frequencies_MHz, 20 * np.log10(abs(h_perturbed)), label='Perturbed Filter', linewidth=2, linestyle='--')

# Mark the lower and upper cutoff frequencies in MHz
plt.axvline(f_low / 1e6, color='r', linestyle='--', label=f'Lower cutoff: {f_low / 1e6:.2f} MHz')
plt.axvline(f_high / 1e6, color='r', linestyle='--', label=f'Upper cutoff: {f_high / 1e6:.2f} MHz')

# Limit the frequency span on the plot
plt.xlim(10, 250)  # Frequency axis in MHz

# Labels and title
plt.title(f'Frequency Response of {n}-th Degree Elliptic Bandpass Filter (with Perturbations)')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True, which="both", ls="--")
plt.legend(loc='best')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.ticker as ticker

# Define filter parameters
n = 4  # Filter order
f_center = 100e6  # Center frequency in Hz (150 MHz)
BW = 100e6  # Bandwidth in Hz (50 MHz)
ripple_dB = 1  # Passband ripple in dB
attenuation_dB = 96  # Stopband attenuation in dB
f_s = 400e6  # 400 MHz sampling frequency
tolerance = 0.01
# Design the elliptic bandpass filter
f_low = f_center - BW / 2  # Lower cutoff frequency
f_high = f_center + BW / 2  # Upper cutoff frequency

# Normalize frequencies by the Nyquist frequency (f_s / 2)
low = f_low / (f_s / 2)
high = f_high / (f_s / 2)

# Ensure normalized frequencies are valid
if not (0 < low < 1) or not (0 < high < 1):
    raise ValueError("Cutoff frequencies must satisfy 0 < f_low, f_high < f_s / 2.")

# Design the elliptic filter using scipy.signal
b, a = signal.ellip(n, ripple_dB, attenuation_dB, [low, high], btype='bandpass', analog=False)

# Apply perturbation to simulate tolerance
def apply_perturbation(coefficients, perturbation_factor=0.01):  # ±1% tolerance
    perturbed_coefficients = []
    for coeff in coefficients:
        perturbation = np.random.uniform(1 - perturbation_factor, 1 + perturbation_factor)
        perturbed_coefficients.append(coeff * perturbation)
    return np.array(perturbed_coefficients)

# Frequency response calculation
w, h = signal.freqz(b, a, worN=8000)
frequencies = w / np.pi * (f_s / 2)  # Convert to Hz
frequencies_MHz = frequencies / 1e6  # Convert to MHz for display

# Initialize variables to track the maximum and minimum responses
max_response = np.zeros_like(frequencies)
min_response = np.ones_like(frequencies) * np.inf  # Set initial min response to infinity

# Perturb the filter 100 times and calculate the frequency response each time
for i in range(1000):
    # Perturb the b and a coefficients
    b_perturbed = apply_perturbation(b, perturbation_factor=tolerance)  # ±1% tolerance
    a_perturbed = apply_perturbation(a, perturbation_factor=tolerance)

    # Calculate the frequency response for the perturbed filter
    _, h_perturbed = signal.freqz(b_perturbed, a_perturbed, worN=8000)

    # Update the max and min responses
    max_response = np.maximum(max_response, np.abs(h_perturbed))
    min_response = np.minimum(min_response, np.abs(h_perturbed))

# Plot frequency response
plt.figure(figsize=(10, 6))

# Plot original frequency response
plt.semilogx(frequencies_MHz, 20 * np.log10(np.abs(h)), label=f'{n}-th degree Elliptic Filter (Original)', linewidth=2)

# Plot the min and max responses
plt.plot(frequencies_MHz, 20 * np.log10(max_response), label='Maximum Response', linestyle='--', color='red')
plt.plot(frequencies_MHz, 20 * np.log10(min_response), label='Minimum Response', linestyle='--', color='blue')

# Mark the lower and upper cutoff frequencies in MHz
plt.axvline(f_low / 1e6, color='r', linestyle='-.', label=f'Lower cutoff: {f_low / 1e6} MHz')
plt.axvline(f_high / 1e6, color='r', linestyle='-.', label=f'Upper cutoff: {f_high / 1e6} MHz')

# Limit the frequency span on the plot
plt.xlim(50, 250)  # Frequency axis in MHz

# Labels and title
plt.title(f'Frequency Response of {n}-th Degree Elliptic Bandpass Filter with Perturbation')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True, which="both", ls="--")
plt.legend(loc='best')

# Use ScalarFormatter to avoid scientific notation on the x-axis
formatter = ticker.ScalarFormatter()
formatter.set_scientific(False)
formatter.set_useOffset(False)

# Apply the formatter to the x-axis
plt.gca().xaxis.set_major_formatter(formatter)

# Customize the x-axis ticks to show in MHz without scientific notation
plt.xticks(np.arange(5, 251, 25))  # Custom ticks in MHz, from 50 MHz to 250 MHz

# Show the plot
plt.tight_layout()
plt.show()
```
