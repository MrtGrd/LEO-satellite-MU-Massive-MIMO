```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'DejaVu Serif'
mpl.rcParams['font.size'] = 10  # Base font size
mpl.rcParams['axes.titlesize'] = 10  # Title font size
mpl.rcParams['axes.labelsize'] = 10  # Axis label font size
mpl.rcParams['xtick.labelsize'] = 10  # X-axis tick label size
mpl.rcParams['ytick.labelsize'] = 10  # Y-axis tick label size
mpl.rcParams['legend.fontsize'] = 10  # Legend font size

# Constants
c = 3e8  # Speed of light in m/s
d_550km = 550000
d_900km = 900000

# Frequency range from 1 GHz to 100 GHz
frequencies = np.logspace(np.log10(1e9), np.log10(100e9), 100)

# Free Space Loss (FSL) calculation
FSL_550km = 20 * np.log10(d_550km) + 20 * np.log10(frequencies) + 20 * np.log10(4 * np.pi / c)
FSL_900km = 20 * np.log10(d_900km) + 20 * np.log10(frequencies) + 20 * np.log10(4 * np.pi / c)

# Plotting
plt.figure(figsize=(8, 6))
plt.semilogx(frequencies, FSL_550km, label="550 km", color='b', linewidth=2)
plt.semilogx(frequencies, FSL_900km, label="900 km", color='r', linewidth=2)
plt.title("Free Space Loss vs Frequency", fontsize=14)
plt.xlabel("Frequency (Hz)", fontsize=12)
plt.ylabel("Free Space Loss (dB)", fontsize=12)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
